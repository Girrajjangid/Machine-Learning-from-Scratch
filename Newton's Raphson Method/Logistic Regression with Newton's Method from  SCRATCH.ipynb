$$
\begin{array}{l}{\text { The gradient of } \ell(\theta) \text { is: }} \\ {\qquad \nabla \ell=\left\langle\begin{array}{c}{\sum_{i=1}^{n}\left(y_{i}-h_{\theta}\left(x_{i}\right)\right) x_{i}} \\ {\sum_{i=1}^{n}\left(y_{i}-h_{\theta}\left(x_{i}\right)\right)}\end{array}\right\rangle}\end{array}
$$
While the Hessian is $\ell(\theta)$
$$
H_{\ell(\hat{\theta})}=\left[\begin{array}{c}{\sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right)  x_ix_i, \sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right) x_i} \\ {\sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right) x_i, \sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right)}\end{array}\right]
$$

$$where\ h_{\theta}(x)=\frac{1}{1+e^{-z}} \text { and } z=\theta_{1} x+\theta_{2}
$$

```python
def sigmoid( theta , data):
    z = np.dot(data ,theta)
    return 1 / (1 + np.exp(-z))

def log_liklihood(data , label , theta ):
    val = (-(np.matmul(label.T , np.log( sigmoid( theta , data))) + 
             (np.matmul( (1 - label).T , np.log(1 - sigmoid(theta,data)) ) ) )/data.shape[0]) 
    return val.reshape(1)

def gradient(data , label ,  theta):  
    return np.dot(data.T , (sigmoid(theta,data) - label))  / data.shape[0]


def hessian_(theta , data):
    a = np.ones([initial_theta.shape[0],data.shape[1]])
    sp = sigmoid( theta , data)                                        
    for i in range(theta.shape[0]):
        for j in range(data.shape[1]):
            a[i][j] = np.sum((sp * (1 - sp)) * data[:,i] * data[:,j])                  
            
    return a

def hessian( theta  , data):  # this and above function both are same I just open them here
    sp = sigmoid( theta , data)                                        
    d00 = np.sum((sp * (1 - sp)) * data[:,0] * data[:,0])                  
    d01 = np.sum((sp * (1 - sp)) * data[:,0] * data[:,1])                  
    d02 = np.sum((sp * (1 - sp)) * data[:,0] * data[:,2])                  

    d10 = np.sum((sp * (1 - sp)) * data[:,1] * data[:,0])                  
    d11 = np.sum((sp * (1 - sp)) * data[:,1] * data[:,1])                  
    d12 = np.sum((sp * (1 - sp)) * data[:,1] * data[:,2])                  

    d20 = np.sum((sp * (1 - sp)) * data[:,2] * data[:,0])                  
    d21 = np.sum((sp * (1 - sp)) * data[:,2] * data[:,1])                  
    d22 = np.sum((sp * (1 - sp)) * data[:,2] * data[:,2])
    
    return np.array([[d00 , d01 , d02] , [d10 ,d11 ,d12] , [d20 , d21 , d22] ])
```

```python
def newtonsMethod():
    global training_data , training_label , fc_N , itera_N
    epsilon = 0.00001
    
    initial_theta = np.ones([training_data.shape[1]+1,1])  # 3 * 1
    data = np.append(np.ones([training_data.shape[0] , 1] , dtype = int) , training_data ,axis = 1 ) # 69 * 3 

    i = 1 
    fc_N = []
    itera_N = []


    while True:
        
        h = hessian( initial_theta , data)
        g = gradient( data , training_label , initial_theta )    

        final_theta = initial_theta -  2 *  np.dot( np.linalg.inv(h) , g)

        initial_cost = log_liklihood(data , training_label , initial_theta)
        final_cost = log_liklihood(data , training_label , final_theta)
    
        if abs(final_cost - initial_cost) < epsilon:
            break
        
        print(i,' old {} new {}'.format(initial_cost , final_cost))
        
        initial_theta = final_theta
        
        fc_N.append(final_cost)
        itera_N.append(i)
        i+=1
        
    return final_theta
```

```python
final_theta_N = newtonsMethod()
```

```python
# Gradient decent

def sigmoidD(theta0, theta , features):
    z = theta0 + np.dot(features ,theta)
    return 1 / (1 + np.exp(-z))

alpha = 0.1
epsilon = 0.00001
initial_theta0 = np.array(1) 
initial_theta = np.ones([training_data.shape[1],1])
i = 0 
fc_GD = []
itera_GD = []
while(True):
    
    del_theta0 = np.sum(sigmoidD(initial_theta0,initial_theta,training_data) - training_label) / training_data.shape[0]
    del_theta = np.dot(training_data.T , (sigmoidD(initial_theta0,initial_theta,training_data) - training_label))
    
    
    final_theta0_GD = initial_theta0 - alpha * del_theta0
    final_theta_GD = initial_theta - alpha * del_theta
    
    initial_cost = (-(np.matmul(training_label.T,np.log(sigmoidD(initial_theta0,initial_theta,training_data))) 
                  + (np.matmul((1-training_label).T,np.log(1-sigmoidD(initial_theta0,initial_theta,training_data)))))
                    /training_data.shape[0])  
    initial_cost = initial_cost.reshape(1)
    
    final_cost = (-(np.matmul(training_label.T,np.log(sigmoidD(final_theta0_GD,final_theta_GD,training_data))) 
                 + (np.matmul((1-training_label).T,np.log(1-sigmoidD(final_theta0_GD,final_theta_GD,training_data)))))
                  /training_data.shape[0])
    final_cost = final_cost.reshape(1)    
    
    if(abs(final_cost - initial_cost) < epsilon):
        break
    print('iteration {} neg log liklihood {}'.format(i,initial_cost))
    
    initial_theta = final_theta_GD
    initial_theta0 = final_theta0_GD
    itera_GD.append(i)
    fc_GD.append(final_cost)
    i+=1
```

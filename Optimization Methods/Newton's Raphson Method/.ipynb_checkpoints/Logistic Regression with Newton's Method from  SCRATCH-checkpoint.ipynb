$$
\begin{array}{l}{\text { The gradient of } \ell(\theta) \text { is: }} \\ {\qquad \nabla \ell=\left\langle\begin{array}{c}{\sum_{i=1}^{n}\left(y_{i}-h_{\theta}\left(x_{i}\right)\right) x_{i}} \\ {\sum_{i=1}^{n}\left(y_{i}-h_{\theta}\left(x_{i}\right)\right)}\end{array}\right\rangle}\end{array}
$$
While the Hessian is $\ell(\theta)$
$$
H_{\ell(\hat{\theta})}=\left[\begin{array}{c}{\sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right)  x_ix_i, \sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right) x_i} \\ {\sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right) x_i, \sum_{i=1}^{n} h_{\theta}\left(x_{i}\right)\left(1-h_{\theta}\left(x_{i}\right)\right)}\end{array}\right]
$$

$$where\ h_{\theta}(x)=\frac{1}{1+e^{-z}} \text { and } z=\theta_{1} x+\theta_{2}
$$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
raw_data = pd.read_csv('marks.txt')
raw_data.columns = ['dogs','cats','labels']
# seperate features and labels
features = raw_data.iloc[:,:-1]  # contains dog and cat features
labels = raw_data.iloc[:,-1]  # contains label 0 and 1
label_1 = features.loc[labels == 1]  # contains dog and cat feature of 1 label
label_0 = features.loc[labels == 0]  # contains dog and cat feature of 0 label
labels = np.array(labels)  
features = np.array(features)
plt.scatter(label_0.iloc[:,0],label_0.iloc[:,1],c = ['blue'],label = 'Dogs')
plt.scatter(label_1.iloc[:,0],label_1.iloc[:,1],c = ['orange'],label = 'Cats')
plt.legend()
plt.show()
```

```python
labels = (labels.reshape(99,1))
# Normalizing  the data
lab1 = label_1/np.max(label_1)
lab0 = label_0/np.max(label_0) 
plt.scatter(lab0.iloc[:,0],lab0.iloc[:,1],c = ['blue'],label = 'Dogs')
plt.scatter(lab1.iloc[:,0],lab1.iloc[:,1],c = ['orange'],label = 'Cats')
plt.legend()
plt.show() # this graph shows that we have getting same graph before and after normalizing the data
div = int(.7*features.shape[0])
# Data seperation
training_data = features[ : div ,: ]
training_data /=np.max(training_data) 
testing_data = features[div:,:]
testing_data /=np.max(testing_data)
training_label = labels[:div]
testing_label = labels[div:]
```

```python
def sigmoid( theta , data):
    z = np.dot(data ,theta)
    return 1 / (1 + np.exp(-z))

def log_liklihood(data , label , theta ):
    val = (-(np.matmul(label.T , np.log( sigmoid( theta , data))) + 
             (np.matmul( (1 - label).T , np.log(1 - sigmoid(theta,data)) ) ) )/data.shape[0]) 
    return val.reshape(1)

def gradient(data , label ,  theta):  
    return np.dot(data.T , (sigmoid(theta,data) - label)) / data.shape[0] 

def hessian_(theta , data):
    a = np.ones([theta.shape[0],data.shape[1]])
    sp = sigmoid( theta , data)                                        
    for i in range(theta.shape[0]):
        for j in range(data.shape[1]):
            a[i][j] = ((np.sum((sp * (1 - sp)) * data[:,i] * data[:,j]))  / data.shape[0])                 
            
    return a

def hessian( theta  , data):  # this and above function both are same I just open them here
    sp = sigmoid( theta , data)                                        
    d00 = np.sum((sp * (1 - sp)) * data[:,0] * data[:,0])                  
    d01 = np.sum((sp * (1 - sp)) * data[:,0] * data[:,1])                  
    d02 = np.sum((sp * (1 - sp)) * data[:,0] * data[:,2])                  

    d10 = np.sum((sp * (1 - sp)) * data[:,1] * data[:,0])                  
    d11 = np.sum((sp * (1 - sp)) * data[:,1] * data[:,1])                  
    d12 = np.sum((sp * (1 - sp)) * data[:,1] * data[:,2])                  

    d20 = np.sum((sp * (1 - sp)) * data[:,2] * data[:,0])                  
    d21 = np.sum((sp * (1 - sp)) * data[:,2] * data[:,1])                  
    d22 = np.sum((sp * (1 - sp)) * data[:,2] * data[:,2])
    
    return np.array([[d00 , d01 , d02] , [d10 ,d11 ,d12] , [d20 , d21 , d22] ])
```

```python
def newtonsMethod():
    global training_data , training_label , fc_N , itera_N
    epsilon = 0.00001
    
    initial_theta = np.ones([training_data.shape[1]+1,1])  # 3 * 1
    data = np.append(np.ones([training_data.shape[0] , 1] , dtype = int) , training_data ,axis = 1 ) # 69 * 3 

    i = 1 
    fc_N = []
    itera_N = []


    while True:
        
        h = hessian_( initial_theta , data)
        g = gradient( data , training_label , initial_theta )    

        final_theta = initial_theta -  np.dot( np.linalg.inv(h) , g)

        initial_cost = log_liklihood(data , training_label , initial_theta)
        final_cost = log_liklihood(data , training_label , final_theta)
    
        if abs(final_cost - initial_cost) < epsilon:
            break
        
        print(i,' old {} new {}'.format(initial_cost , final_cost))
        
        initial_theta = final_theta
        
        fc_N.append(final_cost)
        itera_N.append(i)
        i+=1
        
    return final_theta
```

```python
final_theta_N = newtonsMethod()
```

```python
# Gradient decent

def sigmoidD(theta0, theta , features):
    z = theta0 + np.dot(features ,theta)
    return 1 / (1 + np.exp(-z))

alpha = 0.1
epsilon = 0.00001
initial_theta0 = np.array(1) 
initial_theta = np.ones([training_data.shape[1],1])
i = 0 
fc_GD = []
itera_GD = []
while(True):
    
    del_theta0 = np.sum(sigmoidD(initial_theta0,initial_theta,training_data) - training_label) / training_data.shape[0]
    del_theta = np.dot(training_data.T , (sigmoidD(initial_theta0,initial_theta,training_data) - training_label)) / training_data.shape[0]
    
    
    final_theta0_GD = initial_theta0 - alpha * del_theta0
    final_theta_GD = initial_theta - alpha * del_theta
    
    initial_cost = (-(np.matmul(training_label.T,np.log(sigmoidD(initial_theta0,initial_theta,training_data))) 
                  + (np.matmul((1-training_label).T,np.log(1-sigmoidD(initial_theta0,initial_theta,training_data)))))
                    /training_data.shape[0])  
    initial_cost = initial_cost.reshape(1)
    
    final_cost = (-(np.matmul(training_label.T,np.log(sigmoidD(final_theta0_GD,final_theta_GD,training_data))) 
                 + (np.matmul((1-training_label).T,np.log(1-sigmoidD(final_theta0_GD,final_theta_GD,training_data)))))
                  /training_data.shape[0])
    final_cost = final_cost.reshape(1)    
    
    if(abs(final_cost - initial_cost) < epsilon):
        break
    print('iteration {} neg log liklihood {}'.format(i,initial_cost))
    
    initial_theta = final_theta_GD
    initial_theta0 = final_theta0_GD
    itera_GD.append(i)
    fc_GD.append(final_cost)
    i+=1
```

```python
print('final theta of Newton :' , *final_theta_N , 'with iteration ',itera_N[-1])
print('final theta of gradient decent ' , np.append(final_theta0_GD , final_theta_GD) , 'with iteration ' ,itera_GD[-1])
```

```python
def accuracy(theta_N , theta_GD  ,data , label , threshold):
    z = sigmoidD(theta_N[0] , theta_N[1:] , data)
    for i in range(z.shape[0]):
        if z[i] >= threshold:
            z[i] = int(1)
        else:
            z[i] = 0
    print( "Accuracy or Recall :  taking Newton's method coeff. : " , round(( np.count_nonzero (z == label[:,0])) / label.shape[0]*100 , 2)  ,"%")

    z1 = sigmoidD(theta_GD[0] , theta_GD[1:] , data)
    for i in range(z.shape[0]):
        if z1[i] >= threshold:
            z1[i] = int(1)
        else:
            z1[i] = 0
    print( "Accuracy or Recall :  taking gradient decent coeff. : " , round(( np.count_nonzero (z1 == label[:,0])) / label.shape[0]*100 , 2)  ,"%")
    return z , z1
```

```python
print('Training data')
accuracy(theta_N , theta_GD , training_data , training_label , 0.5)
print('\nTesting data')
accuracy(theta_N , theta_GD , testing_data , testing_label, 0.5)
```

### Basic terms related to Confusion matrix

**True positives (TP):** These are cases in which we predicted yes (employees will leave the organisation), and employees actually leave 

**True negatives (TN):** We predicted no(employees will not leave the organisation) and they don’t leave 

**False positives (FP):** We predicted yes they will leave, but they don’t leave. (Also known as a “Type I error.”) 

**False negatives (FN):** We predicted no they will not leave, but they actually leave (Also known as a “Type II error.”)

Sensitivity and specificity are statistical measures of the performance of a binary classification test:

**Sensitivity/Recall** = TP/(TP + FN). When it’s actually yes, how often does it predict yes? 

**Specificity** = TN/(TN + FP) .When it’s actually no, how often does it predict no?? 

**Precision** = TP/predicted yes. When it predicts yes, how often is it correct?

High Threshold :

High specificity
Low sensitivity
Low Threshold

Low specificity
High sensitivity

The threshold value, t, is often selected based on which errors are better.This would imply that t would be best for no errors but it’s rare to have a model that predicts perfectly,

There are two types of errors that this model can make:
1. where model predicts 1, or poor care, but the actual outcome is 0,

2. Where model predicts 0,or good care, but the actual outcome is 1

```python
def ROC(pred_y):
    from sklearn import metrics as m
    import math
    TPR = []
    FPR = []
    #threshold = np.linspace(.1,.9,6)  # 0  , 0.111 , 0.2222 , 0.333 it break 0 - 1 into ten parts
    threshold = [.999800337, .741326424, .527852556, .247972612,.222412202, .164829506, .00221520725]
    threshold.reverse()
    for k in threshold:
        l = []
        for i in range(pred_y.shape[0]):
            if pred_y[i] > k:
                l.append(1)
            else:
                l.append(0)
        
        confu_matix_N = m.confusion_matrix ( l , testing_label )
        TPR.append((confu_matix_N[1][1] / (confu_matix_N[1][0] + confu_matix_N[1][1])))
        FPR.append((confu_matix_N[0][1] / (confu_matix_N[0][0] + confu_matix_N[0][1])))

    return TPR , FPR
```

```python
pred_y_N = sigmoidD(theta_N[0] , theta_N[1:] , testing_data)
pred_y_GD = sigmoidD(theta_GD[0] , theta_GD[1:] , testing_data)
tpr_N , fpr_N = ROC(pred_y_N)
tpr_GD , fpr_GD = ROC(pred_y_GD)


plt.figure(figsize=(10,10))
plt.plot(fpr_N,tpr_N , label = ' Newton ' )
plt.plot(fpr_GD,tpr_GD , label = 'Gradient Decent')
plt.legend()
plt.title(" Receiver Operator Characteristic ")
plt.show()
```

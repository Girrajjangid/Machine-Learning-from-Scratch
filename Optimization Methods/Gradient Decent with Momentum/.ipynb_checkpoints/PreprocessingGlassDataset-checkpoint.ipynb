```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
raw_data = pd.read_csv('../DataSets/glass.csv')
```

# Preprocessing

```python
raw_data.skew() # most of the parts are skewed
```

```python
from sklearn.model_selection import train_test_split

X = raw_data.iloc[:,:-1]
y = raw_data.iloc[:,-1]

X_train , X_test , y_train , y_test = train_test_split(X,y,
                                                       test_size = 0.3, 
                                                       random_state=42,
                                                       shuffle=True,
                                                       stratify=y)
```

```python
sns.countplot(y_train)
plt.show()
```

```python
sns.countplot(y_test)
plt.show()
```

```python
# Normalize data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_n = sc.fit_transform(X_train)  
X_test_n  = sc.transform(X_test)
```

```python
print(X_train_n.mean(axis=0))
print(X_test_n.mean(axis=0))
```

```python
# remove skewness
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
X_train_pt = pt.fit_transform(X_train)
X_test_pt  = pt.transform(X_test)
```

```python
# Comparing the 3 distributions normal and powertransformer
```

```python
def printDistributions(X_train_n):
    print('\tX_train\t\tX_train_normalize\ttrain_powernormalize')
    for i,feature in enumerate(X_train):
            fig, ax = plt.subplots(1,3,figsize=(8,2.5))    
            ax[0].hist(X_train[feature], color='blue', bins=30, alpha=0.5 )
            ax[0].set_title(str(feature))   
    
            ax[1].hist(X_train_n[:,i], color='red', bins=30, alpha=0.5 )
            ax[1].set_title(str(feature))   
    
            ax[2].hist(X_train_pt[:,i], color='purple', bins=30, alpha=0.5 )
            ax[2].set_title(str(feature))   

            plt.show()

printDistributions(X_train_n)
```

```python
# Remove outliers
# 1. by using quantiles but it divide the data from starting position of distribution 
# 2. Standard daviation based detection
```

```python
from collections import Counter

def outlierDetect(df,no_of_feature_contains = 3):
    '''pass pandas"s dataframe'''
    outlier_indices = []
    for col in df:
        q1 = np.percentile(df[col],25)
        q3 = np.percentile(df[col],75)
        # IQR
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        outlier_index = df[ (df[col] < lower) | (df[col] > upper)].index.tolist()
        outlier_indices.extend(outlier_index)

    counted_outliers = Counter(outlier_indices)    
    indexs = [k for k,v in counted_outliers.items() if v > no_of_feature_contains] 
    print(indexs)
    return indexs
```

```python
outlier_index = outlierDetect(pd.DataFrame(X_train_n))    
```

```python
len(outlier_index)
```

```python
X_train_n = np.delete(X_train_n , outlier_index , axis=0)
X_train_n.shape
```

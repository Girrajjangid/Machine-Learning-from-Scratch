First import required data sets

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as s
```

```python
data = pd.read_csv("../DataSets/binianmalignent.csv")
data.head()
data.info()
```

```python
columns = data.columns
columns
```

```python
sns.heatmap(data.isnull() ,cbar = False) # it show all null values in our data sets with white color
```

```python
# cleaning the data
benign_data = data[data['diagnosis'] == 'B']
benign_data.drop([columns[0],columns[1],columns[-1]] , inplace=True , axis=1)
benign_data.head()
```

Let we assume our data coming from normal distribution
### Steps of Gradient Decent
1. Take gradient of the loss function (MLF)
2. Pick random values for the parameters
3. plug parameter values into the gradient 
4. calculate new_parameter  = old_parameter - slope*learning_rate
5. put new_parameter in function and check with previous values

```python
# we have two parameters here 'sigma' and 'mu' bcz data is of normal distribution
benign_data.head()
```

```python
alpha = 0.001  # learning rate
epsilon = 0.00001 # indicate when to stop iteration
initial_vals = np.random.randn(2,1)  # return 2*1 matrix 
while(True):
    del_mu = -((np.sum((benign_data.iloc[:,0] - initial_vals[0]))) / initial_vals[1]**2) # gradient w.r.t mean
    del_sigma = (benign_data.shape[0]/initial_vals[1]) - ((np.sum((benign_data.iloc[:,0] - initial_vals[0])**2)) / initial_vals[1]**3 )# gradient w.r.t std deviation
    final_vals = initial_vals - alpha * np.array([del_mu,del_sigma]).reshape(2,1)
     
    # now we put new_value into Gaussian function and check weather these parameter are good or not
    fun_old = - np.sum(np.log(s.norm.pdf(benign_data.iloc[:,0],initial_vals[0],initial_vals[1])))
    fun_new = - np.sum(np.log(s.norm.pdf(benign_data.iloc[:,0],final_vals[0],final_vals[1])))
    
    if(abs(fun_new - fun_old) < epsilon):
        break
    print("The initial and final value of Neg Log Likelihood is {} and {}".format(fun_old,fun_new))
    initial_vals = final_vals
```

```python
# Now lets check with original data 
x = benign_data.iloc[:,0]
-np.sum(np.log(s.norm.pdf(x,np.mean(x),np.std(x))))
```

## We get same cost function value with its natural parameter and with gradient decent calculated parameters
# BOOM!!

## Now we calculate for multiclasses

```python
benign_data.head()
```

```python
alpha = 0.01  # learning rate
epsilon = 0.001 # indicate when to stop iteration
sig = np.random.normal(loc=100, size=(len(benign_data.iloc[0, :])))
mu = np.array(benign_data.iloc[0,0:])
initial_val = np.array([mu,sig])
clean_benign_data = np.array(benign_data)

#sigma_cap = np.random.normal(loc=100, size=(len(clean_benign_data[0,:])))
#mu_cap = randColVals(clean_benign_data) 
#initial_val = np.array([sigma_cap,mu_cap])

while(True):
    del_mu = -((np.sum((clean_benign_data - initial_val[0]) , axis=0)) / initial_val[1]**2) # gradient w.r.t mean
    del_sigma = (clean_benign_data.shape[0]/initial_val[1]) - ((np.sum((clean_benign_data - initial_val[0])**2 , axis=0)) / initial_val[1]**3 )# gradient w.r.t std deviation
    final_val = initial_val -  alpha * np.array([del_mu,del_sigma])
    final_val[1] = abs(final_val[1])

    fun_old = - np.sum(np.log(s.norm.pdf(clean_benign_data,initial_val[0],initial_val[1])),axis=0)
    fun_new = - np.sum(np.log(s.norm.pdf(clean_benign_data,final_val[0],final_val[1])),axis=0)
    fun_old_vector = np.sqrt(np.sum(fun_old**2))
    fun_new_vector = np.sqrt(np.sum(fun_new**2))
    if(abs(fun_new_vector - fun_old_vector)) < epsilon:
            break
    print("The initial and final value of Neg Log Likelihood is {} and {}\n".format(fun_old_vector,fun_new_vector))
    initial_val  = final_val
```

```python
c = 0
for i in range(len(final_val[0])):
    if( round(final_val[0][i] , 1) == round(clean_benign_data.mean(axis=0)[i] , 1)):
        c+=1

print("Accuracy : ",round((c/final_val[0].size * 100) , 2 ), "%")
```

```python
class Node:
    def __init__(self, val):
        self.value = val
        self.left_child = None
        self.right_child = None
        self.parent = None
```

```python
class BinaryTree:
    def __init__(self):
        self.root = None
        
    def insert(self , val):
        if self.root == None:
            self.root = Node(val)
        else:
            self._insert(self.root , val)
    
    def _insert(self, cur_node , val):  # cur_node is a reference of cur node or object
        if val < cur_node.value:
            if cur_node.left_child == None:
                cur_node.left_child = Node(val)
                cur_node.left_child.parent = cur_node
            else:
                self._insert(cur_node.left_child , val)
                
        elif val >= cur_node.value:
            if cur_node.right_child == None:
                cur_node.right_child = Node(val)
                cur_node.right_child.parent = cur_node
            else:
                self._insert(cur_node.right_child , val)
                
    def printTree(self):
        if self.root != None :
            self._printTree(self.root)
    
    def _printTree( self , cur_node):
        if cur_node != None:
            self._printTree(cur_node.left_child)
            print((cur_node.value) , end = ' ')
            self._printTree(cur_node.right_child)
            
    def height(self):
        if self.root == None:
            return 0
        else:
            return self._height(self.root , 0)
    
    def _height(self, cur_node , cur_height):
        if cur_node == None : return cur_height
        left_height = self._height(cur_node.left_child , cur_height + 1)
        right_height = self._height(cur_node.right_child , cur_height + 1)
        return max(left_height , right_height)
    
    def search(self, value):
        if self.root != None:
            return self._search(self.root , value)
        else:
            return None
        
    def _search(self , cur_node , val):
        if cur_node.value == val:
            return cur_node
        elif val < cur_node.value and cur_node.left_child!=None: 
            return self._search(cur_node.left_child , val)
        elif val >= cur_node.value and cur_node.right_child!=None:
            return self._search(cur_node.right_child , val)
    def delete(self, val):
        return self.delete_node(self.search(val))  # search()  return the address of that value
    
    def delete_node(self, node):
        if node == None or self.search(node.value) == None:
            print('Node not in the tree')
            return None
        
        def min_value_node(node):
            current = node
            while current.left_child!=None:
                current = current.left_child
            return current

        def num_children(node):
            num_child = 0
            if node.left_child !=None : num_child +=1
            if node.right_child != None : num_child +=1
            return num_child
        
        if node.parent != None:
            node_parent = node.parent
        else:
            node_parent = None
        node_child = num_children(node)
        
        # We have 3 case here
        # case 1 ( node has no children )
        if node_child == 0:
            # added this if statement to check if we delete the root node it would delete the entire tree
            if node_parent != None:
                # remove reference to the node from the parent
                if node_parent.left_child == node: # this is left child
                    node_parent.left_child = None
                else:
                    node_parent.right_child = None
            else:
                self.root = None
        
        # case 2 (node has a single child)
        if node_child == 1:
            if node.left_child != None:
                child = node.left_child
            else:
                child = node.right_child

            if node_parent != None:
                # replace the node to be deleted with its child
                if node_parent.left_child == node:
                    node_parent.left_child = child
                else:
                    node_parent.right_child = child
            else:
                self.root = child
                
            child.parent = node_parent

        # case 3 (NODE has two children)
        if node_child == 2:
            # get the inorder successor of the deleted node
            successor = min_value_node(node.right_child)
            
            # copy the inorder successor's value to the node formerly holding the value we wished to delete
            node.value = successor.value
            self.delete_node(successor)
```

```python
obj = BinaryTree()
l = [4,2,5,1,3,4,6]
for i in l:
    obj.insert(i)
```

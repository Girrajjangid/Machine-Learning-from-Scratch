```python
malignant_tumor_testing_data = pd.DataFrame(raw_data[raw_data['diagnosis'] == 'M']).iloc[199:]
```

```python
malignant_tumor_testing_data.drop(['id','diagnosis','Unnamed: 32'],axis=1,inplace=True)
```

```python
benign_tumor_testing_data = pd.DataFrame(raw_data[raw_data['diagnosis'] == 'B']).iloc[199:]
```

```python
benign_tumor_testing_data.drop(['id','diagnosis','Unnamed: 32'],axis=1,inplace=True)
```

```python
benign_tumor_testing_data
```

```python
malignant_tumor_testing_data
```

```python
correct_count = 0

for i in range(len(malignant_tumor_testing_data)):
    
    tumor_features = np.array(malignant_tumor_testing_data.iloc[i])
    
    likelihood_prob_malignant = sc.multivariate_normal.pdf(tumor_features,
                                                           mean_columns_vector_malignant,
                                                           cov_mat_malignant)
    
    likelihood_prob_benign = sc.multivariate_normal.pdf(tumor_features,
                                                       mean_columns_vector_benign,
                                                       cov_mat_benign)
    
    posterior_prob_malignant = likelihood_prob_malignant/(likelihood_prob_malignant + likelihood_prob_benign)
    
    if posterior_prob_malignant > 0.5:
        
        correct_count += 1
        
    print("The correct counts are",correct_count)
```

```python
pooled_cov_mat = ((len(benign_tumors_training_data)-1)*cov_mat_benign + (len(malignant_tumor_training_data)-1)*cov_mat_malignant)/(398-2)
```

```python
np.linalg.det(pooled_cov_mat)
```

```python
alpha = 0.9

regularized_cov_mat_benign = alpha*pooled_cov_mat + (1-alpha)*cov_mat_benign
```

```python
np.linalg.det(regularized_cov_mat_benign)
```

# Data Normalizing:

```python
def data_normalize(data):
    """Pass any data it will normalize to Standard normalization"""
    a = map(lambda i :(data.iloc[:,i] - np.mean(data.iloc[:,i]))/np.std(data.iloc[:,i]) ,np.arange(len(data.columns)))
    return pd.DataFrame(list(a)).T
```

```python
def heatmap(data):
    """ Pass covariance or correlation with data"""
    plt.figure(figsize=[15,15])
    sns.heatmap(data ,cmap='coolwarm' , annot_kws={'size': 7},
            cbar = True,  square = True, annot=True, fmt= '.1f')
    plt.show()
heatmap(training_data_benign.corr())
```

```python
evalue = 0
def pca_scratch(data , component):
    global evalue
    data = data_normalize(data)
    eign_val , eign_vct = np.linalg.eig(data.cov())
    # now we have to short eign_vals  with  eign_vectors
    sort_index = eign_val.argsort()[ : : -1 ]
    sorted_eign_val = eign_val[sort_index]
    sorted_eign_vct = eign_vct[:,sort_index]
    sorted_labels = np.array(columns_names)[sort_index]

    eign_val_D_space = eign_val[sort_index[0:component]]
    eign_vct_D_space = eign_vct[:,sort_index[0:component]]
    sorted_labels = sorted_labels[:component]

    forming_vector_D = np.dot(data,eign_vct_D_space)

    a = pd.DataFrame(forming_vector_D)
    a.columns = sorted_labels
    evalue = eign_val_D_space
    return a 

data  =  pca_scratch(data_normalize(training_data_benign) , 16)
pd.DataFrame(data)
```

# The probability density function for Univariate Normal is

$$
p\left(x ; \mu, \sigma^{2}\right)=\frac{1}{\sqrt{2 \pi} \sigma} \exp \left(-\frac{1}{2 \sigma^{2}}(x-\mu)^{2}\right)
$$

# The probability density function for Multivariate Normal is
$$
f(x)=\frac{1}{\sqrt{(2 \pi)^{k} \operatorname{det} \Sigma}} \exp \left(-\frac{1}{2}(x-\mu)^{T} \Sigma^{-1}(x-\mu)\right)
$$

where $\mu$ is the mean, $\Sigma$ the covariance matrix, and **k** is the dimension of the space where x takes values.

```python
def jointPDF( data, mean, cov):
    x = np.array(data)
    mean = np.array(mean)
    a = ( 1/(np.sqrt( (((2*np.pi)**20) * (np.linalg.det(cov))) )) ) 
    zero_mean_val = x - mean
    b = np.e ** -((1/2) * np.matmul(zero_mean_val.dot(np.linalg.inv(cov)) , zero_mean_val.T))
    return a * b
```

```python
def probability(data , label_key):
    correct_count = 0 
    for i in range(len(data)):
        
        tumor_features = np.array(data.iloc[i]) # putting data row wise into liklihood
    
        likelihood_prob_malignant = jointPDF(tumor_features ,training_malignent.mean(),training_malignent.cov())
        
        likelihood_prob_benign = jointPDF(tumor_features,training_benign.mean(),training_benign.cov())
    
        posterior_prob_benign = likelihood_prob_benign / (likelihood_prob_malignant + likelihood_prob_benign)
        
        posterior_prob_malignant = likelihood_prob_malignant / (likelihood_prob_malignant + likelihood_prob_benign)
    
        if label_key == 'B':
            if posterior_prob_benign > 0.5 :
                correct_count += 1
        
        elif label_key == 'M':
            if posterior_prob_malignant > 0.5 :
                correct_count +=1
                
    print("\nThe correct counts are",correct_count , " out of ",i)
    p = (correct_count / i )* 100
    print("%.2f"%p,"%" ,end="")
```

```python
probability(testing_benign,"B")   # Testing with Benign data
print("  Accuracy on testing Benign data")
probability(testing_malignent,"M") # Testing with Malignent data
print("  Accuracy on testing Malignent data")
# Testing on whole data
b = (raw_data[raw_data['diagnosis'] == 'B'].drop(['id' , 'diagnosis'],axis=1))  # 357
m = (raw_data[raw_data['diagnosis'] == 'M'].drop(['id' , 'diagnosis'],axis=1))  # 212

probability(pca_scratch(b,c),'B')
print("  Accuracy on Whole Benign data")
probability(pca_scratch(m,c),'M')
print("  Accuracy on Whole malignent data")
```

# Conclusion:

Why accuracy is so minimum ?

1. Maximum likelihood work perfectly when we have large amount of data but here we have only 52 samples for testing.
2. Features are more dependent its difficult to independent them. If we try then may be we lost some information.
3. It is not compulsary that if a distribution has minimum varience then it is less important. In other words, we try to remove those features which has minimum variences or eign values but may be they have more or unique informations. That's why accuracy is low
4. Naive Bayes classifier works when we have categorical data and if we have numerical data then it must be from normal distribution but here not all features are from normal distribution. Some are belongs from Rayle distribution. [left skew]

Naive Bayes Classifier perform well in case of categorical input variables compared to numerical variable(s). For numerical variable, normal distribution is assumed (bell curve, which is a strong assumption)
